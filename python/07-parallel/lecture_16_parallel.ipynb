# Parallel Speedup and Efficiency

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from functools import partial
from ipywidgets import StaticInteract, RangeWidget
```

```python
def parallel_time(n, serial, parallel, overhead):
    return serial + parallel/float(n) + n*overhead/100.

def serial_time(serial, parallel):
    return serial + parallel

def speedup(n, serial, parallel, overhead):
    s = serial_time(serial, parallel)
    p = parallel_time(n, serial, parallel, overhead)
    return s/float(p)
    
num_processors = np.array([1,2,4,8,16,32,64,128])
```

```python
def plot_time(s, o, p=100):
    exetime = partial(parallel_time, serial=s, parallel=p, overhead=o)
    results = map(exetime, num_processors)
    fig, ax = plt.subplots()
    ax.plot(num_processors, results)
    ax.set_xlabel('number of cores')
    ax.set_ylabel('execution time')
    return fig
    
def plot_speedup(s, o, p=100):
    exetime = partial(speedup, serial=s, parallel=p, overhead=o)
    results = map(exetime, num_processors)
    fig, ax = plt.subplots()
    ax.plot(num_processors, results)
    ax.set_xlabel('number of cores')
    ax.set_ylabel('speedup')
    ax.set_xlim(0,140)
    #ax.set_ylim(0,)
    sf = 'serial fraction = {0:.1f}%'.format(s/1.)
    of = 'overhead fraction = {0:.2f}%'.format(o/100.)
    ax.legend([ sf+'\n'+of ], loc='top left')
    return fig    
```

```python
StaticInteract(plot_speedup, s=RangeWidget(0, 10), o=RangeWidget(0, 10))
```
